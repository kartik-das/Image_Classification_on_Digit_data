```python
import matplotlib.image as mpimg

def load_images(file):
    images = []
    for filename in os.listdir(file):
        img = mpimg.imread(os.path.join(file, filename))
        if img is not None:
            images.append(img)
    return images
```

```python
x_data = df.drop("Class", inplace=False, axis=1)
y_data = df["Class"]
```

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#x_train = sc.fit_transform(x_data)
x_train = x_data / 255
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import numpy as np
err = []
for i in range(2,41):
  knn = KNeighborsClassifier(n_neighbors = i,p=2,metric='euclidean')
  knn = knn.fit(x_train, y_data)
  y_pred = knn.predict(x_train)

  miss_count = 0
  y_train = list(y_data)
  for i in range(len(y_pred)):
    if(y_train[i] != y_pred[i]):
      miss_count = miss_count + 1
  error = miss_count / len(y_train)
  err.append(error)
print(err)
```

```python
import matplotlib.pyplot as plt

plt.plot(err,c='r')
plt.xlabel('K')
plt.ylabel('Trainig Error')
plt.title('Training Error vs K')
 # To show the plot
plt.show()
```
